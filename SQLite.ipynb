<h4> Create table

```python
# connecting to the database 
conn = sqlite3.connect("ConSci.db") #creates empty database

# cursor 
cursor = conn.cursor()
```

```python
#daily.info()
```

```python
# creating daily table
cursor.execute("CREATE TABLE daily (date text, \
               ev real, \
               evebit real, \
               evebitda real, \
               marketcap real, \
               pb real, pe real, ps real, ticker text)") 
```

```python
# creating prices table
cursor.execute("CREATE TABLE prices (date text, \
               ticker text, \
               open real, \
               high real, \
               low real, \
               close_split_adjusted real, \
               close_unadjusted real, \
               volume real)") 
```

<h4> Adding data from CSV

```python
# Adding data to daily table

# Adding progress counter
rowsProcessed = 0 

for col in daily.itertuples(): #for row in df
    date = col[1]
    ev = col[2]
    evebit = col[3]
    evebitda = col[4]
    marketcap = col[5]
    pb = col[6]
    pe = col[7]
    ps = col[8]
    ticker = col[9]
    
    # Note, triple quotes indicates multi-line code and you don't have to do the break "\"!
    sql = '''INSERT INTO daily
             (date, ev, evebit, evebitda, marketcap, pb, pe, ps, ticker)
             VALUES
             (:st_date, :st_ev, :st_evebit, :st_evebitda, :st_marketcap, :st_pb, :st_pe, :st_ps, :st_ticker)'''
    # These values are place holders. Tells the SQLite library that something will be substituted here

    # Here, a dictionary has been added of the placeholders and the items to be inserted.
    cursor.execute(sql, {'st_date':date, 'st_ev':ev, \
                             'st_evebit':evebit, 'st_evebitda':evebitda, \
                             'st_marketcap':marketcap, 'st_pb':pb, \
                             'st_pe':pe, 'st_ps':ps, \
                             'st_ticker':ticker})
    
    # Commit. Telling SQLite to save the new data. The data would be lost otherwise.
    conn.commit()
    
    # Progress counter
    rowsProcessed = rowsProcessed + 1
    if rowsProcessed%1000 == 0: 
        print('Progress: {}/{} rows processed'.format(rowsProcessed, len(daily)))
%time
```

```python
# Adding data to prices table
for col in prices.itertuples(): #for row in df
    date = col[1]
    ticker = col[2]
    open = col[3]
    high = col[4]
    low = col[5]
    close_split_adjusted = col[6]
    close_unadjusted = col[7]
    volume = col[8]
    
    # Note, triple quotes indicates multi-line code and you don't have to do the break "\"!
    sql = '''INSERT INTO prices
             (date, ticker, open, high, low, close_split_adjusted, close_unadjusted, volume)
             VALUES
             (:st_date, :st_ticker, :st_open, :st_high, :st_low, :st_close_split_adjusted, :st_close_unadjusted, :st_volume)'''
    # These values are place holders. Tells the SQLite library that something will be substituted here

    # Here, a dictionary has been added of the placeholders and the items to be inserted.
    cursor.execute(sql, {'st_date':date, 'st_ticker':ticker,'st_open':open, 'st_high':high,'st_low':low, \
                         'st_close_split_adjusted':close_split_adjusted,'st_close_unadjusted':close_unadjusted,\
                         'st_volume':volume})
    
    # Commit. Telling SQLite to save the new data. The data would be lost otherwise.
    conn.commit()
%time
```
