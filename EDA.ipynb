```python
# Packages
import pandas as pd
import numpy as np
from IPython.display import display_html # for styler
import matplotlib.pyplot as plt
import seaborn as sns
colors = sns.color_palette('deep')
```

```python
# Datasets
daily = pd.read_csv("daily.csv")
prices = pd.read_csv("prices.csv")
funds = pd.read_csv("funds.csv")
asset_info = pd.read_csv("asset_info.csv")
```

<h3> Dimensions of Data

```python
daily.shape, prices.shape, funds.shape, asset_info.shape
```

<h3> Missing Data

To me, 'roe', 'roic', 'roa', 'ros' seem too important to have missing in our work.

```python
daily_missing = round(daily.isnull().sum()/daily.shape[0]*100,2).to_frame(name = '% missing').sort_values(by = '% missing', ascending = False)
prices_missing = round(prices.isnull().sum()/prices.shape[0]*100,2).to_frame(name = '% missing').sort_values(by = '% missing', ascending = False)
funds_missing = round(funds.isnull().sum()/funds.shape[0]*100,2).to_frame(name = '% missing').sort_values(by = '% missing', ascending = False)
asset_info_missing = round(asset_info.isnull().sum()/asset_info.shape[0]*100,2).to_frame(name = '% missing').sort_values(by = '% missing', ascending = False)
```

```python
# source: https://blog.softhints.com/display-two-pandas-dataframes-side-by-side-jupyter-notebook/
# styler documentation: https://pandas.pydata.org/docs/reference/style.html
space = "\xa0" * 5

df1_styler = funds_missing.style.format(precision = 1)\
                                .set_table_attributes("style='display:inline'")\
                                .set_caption('funds')\
                                .set_table_styles([{'selector': 'caption','props':[('font-weight','700')]}])
df2_styler = daily_missing.style.format(precision = 1)\
                                .set_table_attributes("style='display:inline'")\
                                .set_caption('daily')\
                                .set_table_styles([{'selector': 'caption','props':[('font-weight','700')]}])
df3_styler = asset_info_missing.style.format(precision = 1)\
                                .set_table_attributes("style='display:inline'")\
                                .set_caption('asset_info')\
                                .set_table_styles([{'selector': 'caption','props':[('font-weight','700')]}])
df4_styler = prices_missing.style.format(precision = 1)\
                                .set_table_attributes("style='display:inline'")\
                                .set_caption('prices')\
                                .set_table_styles([{'selector': 'caption','props':[('font-weight','700')]}])

display_html(df1_styler._repr_html_()+space + df2_styler._repr_html_()+space +df3_styler._repr_html_()+space +df4_styler._repr_html_(), raw=True)
#display_html(df1_styler._repr_html_()+df2_styler._repr_html_()+df3_styler._repr_html_()+df4_styler._repr_html_(), raw=True)
```

<h3> Unique Values

```python
rows = []
for col in funds:
  rows.append([col, funds[col].nunique()])
funds_nunique = pd.DataFrame(rows, columns=['variable', 'nunique']).set_index('variable')
```

```python
rows = []
for col in prices:
  rows.append([col, prices[col].nunique()])
prices_nunique = pd.DataFrame(rows, columns=['variable', 'nunique']).set_index('variable')
```

```python
rows = []
for col in asset_info:
  rows.append([col, asset_info[col].nunique()])
asset_info_nunique = pd.DataFrame(rows, columns=['variable', 'nunique']).set_index('variable')
```

```python
# source: https://blog.softhints.com/display-two-pandas-dataframes-side-by-side-jupyter-notebook/
# styler documentation: https://pandas.pydata.org/docs/reference/style.html
space = "\xa0" * 5

df1_styler = funds_nunique.style.format(precision = 1)\
                                .set_table_attributes("style='display:inline'")\
                                .set_caption('funds')\
                                .set_table_styles([{'selector': 'caption','props':[('font-weight','700')]}])
df2_styler = daily_nunique.style.format(precision = 1)\
                                .set_table_attributes("style='display:inline'")\
                                .set_caption('daily')\
                                .set_table_styles([{'selector': 'caption','props':[('font-weight','700')]}])
df3_styler = asset_info_nunique.style.format(precision = 1)\
                                .set_table_attributes("style='display:inline'")\
                                .set_caption('asset_info')\
                                .set_table_styles([{'selector': 'caption','props':[('font-weight','700')]}])
df4_styler = prices_nunique.style.format(precision = 1)\
                                .set_table_attributes("style='display:inline'")\
                                .set_caption('prices')\
                                .set_table_styles([{'selector': 'caption','props':[('font-weight','700')]}])

display_html(df1_styler._repr_html_()+space + df2_styler._repr_html_()+space +df3_styler._repr_html_()+space +df4_styler._repr_html_(), raw=True)
#display_html(df1_styler._repr_html_()+df2_styler._repr_html_()+df3_styler._repr_html_()+df4_styler._repr_html_(), raw=True)
```

<h3> Market Analysis

<b> How-to-read: </b> Assuming asset_info has captured a complete look at the market, 20% of tickers are in the Industrials Sector.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 15))

# Plot 1 
count_by_sector = asset_info.sector.value_counts().to_frame(name = 'ct')
data = count_by_sector.ct
labels = count_by_sector.index
ax1.pie(data, labels=labels, colors=colors, autopct='%.0f%%', textprops={'fontsize': 12}) #radius=1.5
ax1.set_title('Sectors', fontweight=700)

# Plot 2 
top = 9
count_by_fama_industry = asset_info.fama_industry.value_counts().to_frame(name = 'ct')
count_by_fama_industry_top = count_by_fama_industry[:top].copy()
new_row = pd.DataFrame(data = {'index' : ['others'], 'ct':[count_by_fama_industry['ct'][top:].sum()]}).set_index('index')
count_by_fama_industry2 = pd.concat([count_by_fama_industry_top, new_row])
data = count_by_fama_industry2.ct
labels = count_by_fama_industry2.index
ax2.pie(data, labels=labels, colors=colors, autopct='%.0f%%',  textprops={'fontsize': 12}) #radius=1.5
ax2.set_title('Fama_Industries', fontweight=700)

# Plotting
fig.subplots_adjust(wspace=.5)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 15))

# Plot 1 
count_by_exchange = asset_info.exchange.value_counts().to_frame(name = 'ct')
data = count_by_exchange.ct
labels = count_by_exchange.index
ax1.pie(data, labels=labels, colors=colors, autopct='%.0f%%', textprops={'fontsize': 12}) #radius=1.5
ax1.set_title('Exchanges', fontweight=700)

# Plot 2
top = 9
count_by_category = asset_info.category.value_counts().to_frame(name = 'ct')
count_by_category_top = count_by_category[:top].copy()
new_row = pd.DataFrame(data = {'index' : ['others'], 'ct':[count_by_category['ct'][top:].sum()]}).set_index('index')
count_by_category2 = pd.concat([count_by_category_top, new_row])
data = count_by_category2.ct
labels = count_by_category2.index
ax2.pie(data, labels=labels, colors=colors, autopct='%.0f%%',  textprops={'fontsize': 12}) #radius=1.5
ax2.set_title('Categories', fontweight=700)

# Plotting
fig.subplots_adjust(wspace=.5)
plt.show()
```
